```python
import numpy as np
import matplotlib.pyplot as plt 
import os
from prepare_data import prepare_input
```

```python
data_folder = 'data/size60/seed20/'
X = np.fromfile(data_folder + 'exp_number0.npy')
Cost, p, q, s = prepare_input(N=10,size=60,seed=20)
x0 = np.tensordot(p, np.tensordot(q, s, 0), 0)
opt = (X * Cost.reshape(-1)).sum().item()

```

```python
#input ground truth

algs={'M':1e-5,'A':1,'B':1,'C':1}
keys=['Obj_list','runtime','distance']
max_iter = 20000

plt.figure()
for key in keys:
    for alg in algs:
        #input A
        

        output_folder = 'output_60/'+'max_iter-'+f'{max_iter}'
        save_folder=os.path.join(output_folder,f'{alg}-{max_iter}')

        
    
        ep=algs[alg]
        name=os.path.join(save_folder,f'{alg}_{key}{ep}'+'.npy')
        if os.path.exists(name):
            key_value = np.load(name)
        else:
            print(f'{name} not exist')
        if key=='Obj_list':
            plt.plot(abs(key_value-opt),label=f'{alg}')
        else:
            plt.plot(key_value,label=f'{alg}')    


    plt.ylabel(f'{key}')
    plt.legend()
    
    plt.yscale('log')
    plt.show()
```

```python
#input ground truth


algs={'M':[-5],'A':range(-3,2)}
keys=['Obj_list','runtime','distance']
max_iter = 20000

plt.figure()
for key in keys:
    for alg in algs:
        #input A
        
        for i in algs[alg]:
            ep=10**i
            output_folder = 'output_60/'+'max_iter-'+f'{max_iter}'
            save_folder=os.path.join(output_folder,f'{alg}-{max_iter}')

            
            name=os.path.join(save_folder,f'{alg}_{key}{ep}'+'.npy')
            if os.path.exists(name):
                key_value = np.load(name)

            else:
                print(f'{name} not exist')
            if key=='Obj_list':
                plt.plot(abs(key_value-opt),label=f'{alg}-{ep}')
            else:
                plt.plot(key_value,label=f'{alg}-{ep}')    


    plt.ylabel(f'{key}')
    plt.legend(loc='upper left', bbox_to_anchor=(-0.1, 1))
    
    plt.yscale('log')
    plt.show()
```
